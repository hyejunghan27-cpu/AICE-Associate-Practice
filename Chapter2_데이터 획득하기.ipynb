# SECTION 01. 파이썬 데이터 분석 라이브러리 이용하기

## 1. 넘파이 이용하기

```python
# 넘파이(Numpy)불러오기
import numpy as np

# 넘파이(Numpy) 데이터 만들기
# np.array([])
a = np.array([1,2,3,4,5])
b = np.array(['대한민국', '포르투칼', '가나', '우르과이'])
c = np.array([1,2,'대한민국', '포르투칼'])

#출력하기
print (a)
print (b)
print (c)
```

## 2. 판다스 이용하기

```python
# 판다스(Pandas)불러오기
import pandas as pd

# 시리즈(Series)생성하기
# pd.Series([], index=[], name='')
s=pd.Series(['대한민국','포르투칼','가나','우르과이'], index=['가','나','다','라'], name='2022년 카타르 월드컵 H조')

# 출력하기
print(s)
```

```python
#딕셔너리(Dictionary)형으로 데이터프레임(DataFrame) 생성하기
# pd.DataFrame({"a": [1,2,3], "b": [4,5,6]})
a1 = pd.DataFrame({"a" : [1,2,3], "b" : [4,5,6], "c" : [7,8,9]})
a1
```

```python
# 리스트(LIST)형으로 데이터프레임(DataFrame) 생성하기
# 인덱스명 따로 주기
# pd.DataFrame([[1,2,3], [4,5,6], [7,8,9]], ["a", "b", "c"])
a2 = pd.DataFrame([[1,2,3], [4,5,6], [7,8,9]], ["a","b","c"])
a2
```

```python
# 대/소문자 미 구분시 에러 메세지
# pd.Series라고 해야지 에러 안남.
s=pd.series(['대한민국','포르투칼','가나','우르과이'], index=['가','나','다','라'], name='2022년 카타르 월드컵 H조')
print(s)
```

```python
# 대/소문자를 구별하지 않는 경우 에러 메세지
# pd.DataFrame이라고 해야지 에러 안남
a1 = pd.dataframe({"a" : [1,2,3], "b" : [4,5,6], "c" : [7,8,9]})
a1
```
